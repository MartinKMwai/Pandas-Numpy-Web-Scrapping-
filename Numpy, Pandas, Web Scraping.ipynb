```python
#Q2 Part (i)

import requests
from bs4 import BeautifulSoup
import pandas as pd
URL = "https://www.hkbea.com/cgi-bin/rate_ttfx.jsp?language=en"

r = requests.get(URL)
soup = BeautifulSoup (r.content, 'lxml')
print (soup.prettify())    
```

## Part ii

```python
table = soup.find ('table')
```

##### answer

```python
headers = []
for i in table.find_all('th'):
    title = i.text
    headers.append(title)
df= pd.DataFrame (columns = headers)   

for row in table.find_all ('tr')[1:]:
    data = row.find_all ('td')
    row_data = [td.text.strip() for td in data]
    length = len(df)
    df.loc[length] = row_data
    
newtable  = df.head(12)
newtable
```

```python

#establish what data in what columns is to be used
exchange_table = pd.DataFrame(columns = ['Currency', 'Unit', 'Bank Buy', 'Bank Sell'])

# create function to calculate how much the HKD is sold for in a the given currency in the user input
def selling(s, unit, rate):
    return ((unit*s)/rate)
#create function to calculate how much the HKD is bought for in the currency provided in user input
def buying(b, unit, rate): 
  return((rate*b)/unit)
# make the datatypes floats so as to perform functions with them
exchange_table['Unit'] = pd.to_numeric(exchange_table['Unit'], errors='coerce') #return Nan rather than raise exceptions as is default
exchange_table['Bank Buy'] = pd.to_numeric(exchange_table['Bank Buy'], errors='coerce')#return Nan rather than raise exceptions
exchange_table['Bank Sell'] = pd.to_numeric(exchange_table['Bank Sell'], errors='coerce')#return Nan rather than raise exceptions
exchange_table.at[3, 'Unit'] = 100000

show = np.where(exchange_table['Currency'].str.contains(currency, case=False)) #inspecting series
show = int(show[0])
unit = exchange_table.iloc[show]['Unit']
amount = int(amount) #explicit value conversion


if (buy_or_sell == 's'): #looking to sell HKD
  rate = exchange_table.iloc[show]['Bank Buy']
  output = buying(amount, unit, rate)
  print(amount,"HKG", "=", output, currency )
  

else: #while looking to buy HKD instead i.e, buy_or_sell is'b'
  rate = exchange_table.iloc[show]['Bank Sell'] 
  output = selling(amount, unit, rate)
  print(amount, currency, "=", output, "HKG" ) 
```
